```python
epochs = 100
```

```python
import pandas as pd

df = pd.read_csv('./csvs/DN121_MSC.csv')

for col_index, column_name in enumerate(df.columns):
    print(f"Column Index: {col_index}, Column Name: '{column_name}'")
```

```python
import matplotlib.pyplot as plt

num_rows = df.shape[0]
print("Num of rows:",num_rows)
```

```python
def plot(epochs,columns=[],ylabel="",title=""):
    start_row = 0
    end_row = epochs-1

    columns_to_plot = columns

    column_names = [df.columns[col_index] for col_index in columns_to_plot]

    for col_index, column_name in zip(columns_to_plot, column_names):
        plt.plot(df.index[start_row:end_row+1], df.iloc[start_row:end_row+1, col_index], label=f'{column_name}')

    x_ticks = list(range(0, epochs+1,10))
    plt.xticks(x_ticks)
    
    plt.xlabel('Epochs')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.savefig(ylabel+".jpg")
    plt.show()
```

```python
plot(epochs,[0],"train_loss")
```

```python
plot(epochs,[1],"train_accuracy")
```

```python
plot(epochs,[2],"val_loss")
```

```python
plot(epochs,[3],"val_accuracy")
```
